## Label Encoder

#### using for loop

```python
df2.head()
```

```python
df2.columns
```

```python
#for loop
for col in df2.columns:
    if df2[col].dtype == np.number:
        continue
    else:
        df2[col] = le.fit_transform(df2[col])
```

```python
df2.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    df3[col] = le.fit_transform(df3[col])
```

#### using for loop

```python
for col in df6.columns:
    if is_numeric_dtype(df6[col]):
        continue
    one_hot = pd.get_dummies(df6[col], drop_first=True, prefix='city')
    df6.drop(df6[[col]], axis=1, inplace=True) # will drop the corresponding column
    df6 = pd.concat([df6, one_hot], axis=1)
```

## Ordinal

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
cities = df7.Area.values
```

```python
cities
```

```python
cities = df8.Area.unique()
```

```python
cities
```

```python
ordinal = OrdinalEncoder(categories=[cities])
```

```python
ordinal
```

```python
encoded = ordinal.fit_transform(df7[['Area']])
```

```python
encoded
```

```python
pd.DataFrame(encoded, columns=['new_Area'])
```

```python
df7['Area'] = pd.DataFrame(encoded, columns=['new_Area'])
```
